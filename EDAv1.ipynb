1. Memperbaiki atribut pada setiap kolom dari uppercase menjadi title

Mengubah datatype
1. Mengubah datatype tahun dari int menjadi datetime

Transformasi data
1. Mengubah atribut Makanan pada kolom kategori_usaha menjadi Makanan & Minuman
2. Mengubah atribut Minuman pada kolom kategori_usaha menjadi Makanan & Minuman
3. Mengubah atribut Batik pada kolom kategori_usaha menjadi Fashion


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df_umkm = pd.read_excel("diskuk-od_17371_jml_ush_mikro_kecil_menengah_umkm__kabupatenkota_kateg_data.xlsx")
df_umkm.head()
```

```python
df_kemiskinan = pd.read_csv("bps-od_20003_angka_garis_kemiskinan_berdasarkan_kabupatenkota_data.csv")
df_kemiskinan.head()
```

## Memperbaiki nama atribut di tiap dataframe

```python
df_umkm['nama_provinsi'] = df_umkm['nama_provinsi'].str.title()
df_umkm['nama_kabupaten_kota'] = df_umkm['nama_kabupaten_kota'].str.title()
df_umkm['kategori_usaha'] = df_umkm['kategori_usaha'].str.title()
df_umkm['satuan'] = df_umkm['satuan'].str.title()
df_umkm.head()
```

```python
df_kemiskinan['nama_provinsi'] = df_kemiskinan['nama_provinsi'].str.title()
df_kemiskinan['nama_kabupaten_kota'] = df_kemiskinan['nama_kabupaten_kota'].str.title()
df_kemiskinan['satuan'] = df_kemiskinan['satuan'].str.title()
df_kemiskinan.head()
```

## Eksplorasi dataframe UMKM dan Kemiskinan

```python
df_umkm.info()
```

```python
df_kemiskinan.info()
```

### Mengganti datatype atribut

```python
df_umkm["tahun"] = pd.to_datetime(df_umkm["tahun"], format="%Y")
df_umkm.info()
```

```python
df_kemiskinan["tahun"] = pd.to_datetime(df_kemiskinan["tahun"], format="%Y")
df_kemiskinan.info()
```

## Transformasi Data UMKM

```python
df_umkm.kategori_usaha = df_umkm.kategori_usaha.replace({"Makanan": "Makanan & Minuman", "Minuman": "Makanan & Minuman", "Batik": "Fashion"})
```

### Penggabungan kategori usaha pada dataframe UMKM

```python
df_umkm = df_umkm.groupby(['nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'kategori_usaha', 'satuan', 'tahun'])['jumlah_umkm'].sum().reset_index()
df_umkm = df_umkm[['nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'kategori_usaha','jumlah_umkm', 'satuan', 'tahun']]
df_umkm.head()
```

### Pembatasan DataFrame Kemiskinan dari tahun 2017 sampai 2022

```python
df_kemiskinan = df_kemiskinan[(df_kemiskinan['tahun']>'2016-01-01') & (df_kemiskinan['tahun']<'2022-01-01')]
df_kemiskinan = df_kemiskinan[['kode_provinsi', 'nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'angka_garis_kemiskinan', 'satuan', 'tahun']]
```

## Merge Kedua DataFrame

```python
df2 = pd.merge(df_umkm, df_kemiskinan,how="left", on=["kode_kabupaten_kota", "tahun"])
df2.head()
```

### Penghapusan nama atribut yang terduplikasi

```python
df2 = df2.drop(columns=['nama_provinsi_y', 'nama_kabupaten_kota_y', 'satuan_x', 'satuan_y'])
df2 = df2.rename(columns={'nama_provinsi_x': 'nama_provinsi', 'nama_kabupaten_kota_x': 'nama_kabupaten_kota'})
df2["kode_provinsi"] = df2["kode_provinsi"].astype("category")
df2["kode_kabupaten_kota"] = df2["kode_kabupaten_kota"].astype("category")
df2.head()
```

```python
df2 = df2[['kode_provinsi', 'nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'kategori_usaha', 'jumlah_umkm', 'angka_garis_kemiskinan', 'tahun']]
df2.head()
```

# Visualisasi

```python
df3 = df2.groupby(['nama_kabupaten_kota', 'angka_garis_kemiskinan', 'tahun'])['jumlah_umkm'].sum().reset_index()
df2017 = df3[df3['tahun'] == '2017-01-01']
df2018 = df3[df3['tahun'] == '2018-01-01']
df2019 = df3[df3['tahun'] == '2019-01-01']
df2020 = df3[df3['tahun'] == '2020-01-01']
df2021 = df3[df3['tahun'] == '2021-01-01']
```

Boxplot untuk menampilkan atribut kategori usaha berdasarkan jumlah umkm

```python
df2.boxplot(by ='kategori_usaha', column =['jumlah_umkm'], grid = False)
plt.title("Jumlah UMKM")
plt.xlabel('Kategori Usaha')
plt.xticks(rotation=90)
plt.show()
```

Catplot untuk menampilkan angka kemiskinan pada setiap daerah per tahunnya

```python
import warnings
warnings.filterwarnings('ignore')
sns.catplot(x="nama_kabupaten_kota", y="angka_garis_kemiskinan",hue="tahun", data=df2, height=10)
plt.xticks(rotation=90)
plt.show()
```

Membandingkan rata-rata jumlah UMKM dengan angka kemiskinan pada tahun 2017


```python
df_groups2017 = df2017.groupby(['nama_kabupaten_kota'])[['jumlah_umkm', 'angka_garis_kemiskinan']].sum()
df_groups2017.plot(kind='bar')
plt.xlabel('2017')
max_y = np.ceil(df_groups2017[['jumlah_umkm', 'angka_garis_kemiskinan']].max().max() / 50000) * 50000
plt.yticks(np.arange(0, max_y + 1, 50000))
plt.show()
```
